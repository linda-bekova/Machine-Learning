## SVM

Jádrové funkce, co jsem si zvolila pro analýzu jsou linear a sigmoid.

````python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.drop(columns=['label']), data.label, test_size=0.2, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state=42) 
````

### Linear

````python
from sklearn import svm
import sklearn.metrics as metrics
linear = svm.SVC(kernel='linear') # podle dokumentace se defaultní hodnota C rovná 1
linear.fit(x_train, y_train)
train = metrics.accuracy_score(y_train, linear.predict(x_train))

original_linear = metrics.accuracy_score(y_val, linear.predict(x_val))
print('accuracy score (val): {0:.6f}'.format(original_linear))
````

````python
print('accuracy score (test): {0:.6f}'.format(metrics.accuracy_score(y_test, linear.predict(x_test))))
````

Protože regulační parametr C kontroluje počet chyb, které jsme ochotni tolerovat na trénovací množině, budeme testovat, jak ovnlivňuje hodnotu úspěšnosti modelu. Abychom zjistili, která hodnota C je nejlepší, budeme testovat nejprve vyšší hodnoty C, které by mělo povolit menší množství špatně klasifikovaných bodů a poté testovat, jak se mění úspěšnost modelu pro menší hodnoty C.

````python
# vyšší hodnoty C
val = []
train = []
parameters = range(1,100)

for parameter in parameters:
    dt = svm.SVC(C=parameter, kernel='linear')
    dt.fit(x_train, y_train)
    train.append(metrics.accuracy_score(y_train, dt.predict(x_train)))
    val.append(metrics.accuracy_score(y_val, dt.predict(x_val)))
````

````python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,6))
plt.plot(parameters, train,'or-')
plt.plot(parameters, val,'ob-')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend(['train', 'validation'])
````

````python
optimal_larger = np.argmax(val) + 1
optimal_larger
````

Z doposud testovaných hodnot má nejlepší úspěšnost na validačních datech C=1, což už jsme již testovali.

````python
val = []
train = []
parameters = np.arange(0.01, 1.1, 0.01)

for parameter in parameters:
    dt = svm.SVC(C=parameter, kernel='linear')
    dt.fit(x_train, y_train)
    train.append(metrics.accuracy_score(y_train, dt.predict(x_train)))
    val.append(metrics.accuracy_score(y_val, dt.predict(x_val)))
````

````python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,6))
plt.plot(parameters, train,'or-')
plt.plot(parameters, val,'ob-')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend(['train', 'validation'])
````

````python
optimal_smaller = np.argmax(val) * 0.1 + 0.1
optimal_smaller
````

````python
linear = svm.SVC(C=optimal_larger, kernel='linear')
linear.fit(x_train, y_train)
linear_larger = metrics.accuracy_score(y_val, linear.predict(x_val))
print('accuracy score (val): {0:.6f}'.format(linear_larger))
````

````python
linear = svm.SVC(C=optimal_smaller, kernel='linear')
linear.fit(x_train, y_train)
linear_smaller = metrics.accuracy_score(y_val, linear.predict(x_val))
print('accuracy score (val): {0:.6f}'.format(linear_smaller))
````

Jak C=1 tak i C=0.1 dochází ke stejné úspěšnosti. Což znamená, že jsme netestovali dostatečně malé hodnoty C.

````python
val = []
train = []
parameters = np.arange(0.00000001, 0.00001, 0.00000001)

for parameter in parameters:
    dt = svm.SVC(C=parameter, kernel='linear')
    dt.fit(x_train, y_train)
    train.append(metrics.accuracy_score(y_train, dt.predict(x_train)))
    val.append(metrics.accuracy_score(y_val, dt.predict(x_val)))
````

````python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,6))
plt.plot(parameters, train,'or-')
plt.plot(parameters, val,'ob-')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend(['train', 'validation'])
````

````python
optimal_linear = np.argmax(val) * 0.00000001 + 0.00000001
optimal_linear
````

V případě, že drasticky snížíme C hodnotu, dojdeme k o něco lepší úspěšnosti. Budeme tedy považovat tuto hodnotu za optimální.

````python
linear = svm.SVC(C=optimal_linear, kernel='linear')
linear.fit(x_train, y_train)

optimal_linear_val = metrics.accuracy_score(y_val, linear.predict(x_val))
print('accuracy score (val): {0:.6f}'.format(optimal_linear_val))

optimal_linear_test = metrics.accuracy_score(y_test, linear.predict(x_test))
print('accuracy score (val): {0:.6f}'.format(optimal_linear_test))
````

Díky tomu je naivní Bayes vhodný pro tuto úlohu. Kvůli určité rezistenci proti problémům s dimenzionalitou a zároveň, protože potřebuje poměrně malé množství dat na trénovaní.

## LDA

LDA dokáže dobře pracovat i s relativně malým množstvím dat, což se nám v tomto případě hodí. Problém s LDA je, že někdy nemusí dávat dostatečně dobrou accuracy, v případě, že se týká o dataset s vysokou dimenzí.

````python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit_transform(x_train, y_train)

original_lda = metrics.accuracy_score(y_val, lda.predict(x_val))
print('accuracy score (val): {0:.6f}'.format(original_lda))
````

````python
original_lda_test = metrics.accuracy_score(y_test, lda.predict(x_test))
print('accuracy score (test): {0:.6f}'.format(original_lda_test))
````

Jak můžeme vidět, v tomto případě neměl model problém s dimenzí a je schopna dělat dobrá vyhodnocení testovací množiny.

## PCA

V této sekci použijeme PCA za účelem redukci dimenze a následného testování předchozích algoritmů. Snažíme se najít, jaká dimenze je nejlepší pro jaký algoritmus (SVM, NB, LDA).

````python
from sklearn.decomposition import PCA

val_linear = []
val_sigmoid = []
val_nb = []
val_lda = []
parameters = range(1,481)

for parameter in parameters:
    pca = PCA(n_components=parameter)
    x_train_reduced = pca.fit_transform(x_train_standard)
    x_val_reduced = pca.transform(x_val_standard)
    
    linear = svm.SVC(C=optimal_linear, kernel='linear')
    linear.fit(x_train_reduced, y_train)
    val_linear.append(linear.score(x_val_reduced, y_val))
    
    sigmoid = svm.SVC(C=optimal_sigmoid, kernel='sigmoid')
    sigmoid.fit(x_train_reduced, y_train)
    val_sigmoid.append(sigmoid.score(x_val_reduced, y_val))
    
    nb = GaussianNB()
    nb.fit(x_train_reduced, y_train)
    val_nb.append(nb.score(x_val_reduced, y_val))
    
    lda = LinearDiscriminantAnalysis()
    lda.fit(x_train_reduced, y_train)
    val_lda.append(lda.score(x_val_reduced, y_val))
````

````python
model_linear = svm.SVC(C=optimal_linear, kernel='linear')
model_linear.fit(x_train,y_train)
full_acc_linear = model_linear.score(x_val, y_val)

model_sigmoid = svm.SVC(C=optimal_sigmoid, kernel='sigmoid')
model_sigmoid.fit(x_train, y_train)
full_acc_sigmoid = model_sigmoid.score(x_val, y_val)
    
model_nb = GaussianNB()
model_nb.fit(x_train, y_train)
full_acc_nb = model_nb.score(x_val, y_val)
    
model_lda = LinearDiscriminantAnalysis()
model_lda.fit(x_train, y_train)
full_acc_lda = model_lda.score(x_val, y_val)
````

### SVM linear

````python
val_linear = np.array(val_linear)
plt.subplots(1,1, figsize=(15, 8))
ns = plt.scatter(range(1,481), val_linear, c='red')
plt.title(u"Přesnost jako funkce počtu hlavních komponent")
plt.xlabel(u'# hlavních komponent')
plt.ylabel('přesnost')
plt.plot([0, 481], [full_acc_linear, full_acc_linear],'b-')
plt.show()

print(f"Maximální přesnost = {val_linear.max()} pro projekci dimenze {np.argmax(val_linear) + 1}.")
````

````python
pca = PCA(n_components=np.argmax(val_linear) + 1)
x_train_reduced = pca.fit_transform(x_train_standard)
x_val_reduced = pca.transform(x_val_standard)
x_test_reduced = pca.transform(x_test_standard)
    
linear = svm.SVC(C=optimal_linear, kernel='linear')
linear.fit(x_train_reduced, y_train)
print('accuracy score (test): {0:.6f}'.format(metrics.accuracy_score(y_test, linear.predict(x_test_reduced))))
````

### SVM sigmoid

````python
val_sigmoid = np.array(val_sigmoid)
plt.subplots(1,1, figsize=(15, 8))
ns = plt.scatter(range(1,481), val_sigmoid, c='red')
plt.title(u"Přesnost jako funkce počtu hlavních komponent")
plt.xlabel(u'# hlavních komponent')
plt.ylabel('přesnost')
plt.plot([0, 481], [full_acc_sigmoid, full_acc_sigmoid],'b-')
plt.show()

print(f"Maximální přesnost = {val_sigmoid.max()} pro projekci dimenze {np.argmax(val_sigmoid) + 1}.")
````

## LLE

LLE je schopné pracovat i s datasety s vysokými dimenzemi. V této sekci budeme testovat SVM, NB a LDA po redukci dimenze pomocí LLE. A zároveň zkoumat při jakým dimenzích vyhodnocují modely nejlepší výsledky.

````python
from sklearn.manifold import LocallyLinearEmbedding
val_linear = []
val_sigmoid = []
val_nb = []
val_lda = []
parameters = range(1,x_train.shape[1])

for parameter in parameters:
    lle = LocallyLinearEmbedding(n_components = parameter, n_neighbors=5)
    x_train_reduced = lle.fit_transform(x_train_standard)
    x_val_reduced = lle.transform(x_val_standard)
    
    linear = svm.SVC(C=optimal_linear, kernel='linear')
    linear.fit(x_train_reduced, y_train)
    val_linear.append(linear.score(x_val_reduced, y_val))
    
    sigmoid = svm.SVC(C=optimal_sigmoid, kernel='sigmoid')
    sigmoid.fit(x_train_reduced, y_train)
    val_sigmoid.append(sigmoid.score(x_val_reduced, y_val))
    
    nb = GaussianNB()
    nb.fit(x_train_reduced, y_train)
    val_nb.append(nb.score(x_val_reduced, y_val))
    
    lda = LinearDiscriminantAnalysis()
    lda.fit(x_train_reduced, y_train)
    val_lda.append(lda.score(x_val_reduced, y_val))
````

````python
model_linear = svm.SVC(kernel='linear')
model_linear.fit(x_train,y_train)
full_acc_linear = model_linear.score(x_val, y_val)

model_sigmoid = svm.SVC(kernel='sigmoid')
model_sigmoid.fit(x_train, y_train)
full_acc_sigmoid = model_sigmoid.score(x_val, y_val)
    
model_nb = GaussianNB()
model_nb.fit(x_train, y_train)
full_acc_nb = model_nb.score(x_val, y_val)
    
model_lda = LinearDiscriminantAnalysis()
model_lda.fit(x_train, y_train)
full_acc_lda = model_lda.score(x_val, y_val)
````
